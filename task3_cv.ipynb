```python
df1 = pd.read_csv('/mnt/Coding/STT AI/CS203_Lab_3/CV_Praneel.csv')
df2 = pd.read_csv('/mnt/Coding/STT AI/CS203_Lab_3/CV_samarth.csv')
```

```python
final_df = pd.DataFrame(columns=["Image_ID", "Annotator_1", "Annotator2"])
final_df["Image_ID"] = df1["id"]
final_df["Annotator_1"] = df1["choice"]
final_df["Annotator2"] = df2["choice"]
final_df
```

```python
import pandas as pd
from statsmodels.stats.inter_rater import fleiss_kappa

# Load the annotations from a CSV file
# CSV format: Image_ID, Annotator_1, Annotator_2


# Define the categories
categories = ["TRUCK", "NO TRUCK"]

# Initialize the counts matrix
counts = []

# Reformat the data for Fleiss' kappa
for _, row in final_df.iterrows():
    counts.append([sum(row == category) for category in categories])

# Convert to a DataFrame for better visualization (optional)
counts_df = pd.DataFrame(counts, columns=categories)

# Calculate Fleiss' Kappa
kappa = fleiss_kappa(counts_df.values, method='fleiss')
print(f"Fleiss' Kappa: {kappa}")
```
